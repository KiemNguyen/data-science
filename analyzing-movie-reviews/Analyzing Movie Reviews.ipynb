## 3. Mean, Median, And Standard Deviation

```python
fandango_mean = movies["Fandango_Stars"].mean()
metacritic_mean = movies["Metacritic_norm_round"].mean()

fandango_median = np.median(movies["Fandango_Stars"])
metacritic_median = np.median(movies["Metacritic_norm_round"])

fandango_std = np.std(movies["Fandango_Stars"])
metacritic_std = np.std(movies["Metacritic_norm_round"])

print(fandango_mean)
print(metacritic_mean)
print(fandango_median)
print(metacritic_median)
print(fandango_std)
print(metacritic_std)
```

## Fandago vs Metacritic

Fandago appears to inflate their rating. The median metacritic score is higher than the mean metacritic score because some low review rows drag down the median. The median fandango score is lower than the mean fandango score because some high review drag up the median. Fandango ratings in general appear to be higher than metacritic ratings.

## 4. Scatter Plots

Scatterplot that compares the Fandango_Stars column to the Metacritic_norm_round column.

```python
plt.scatter(movies["Fandango_Stars"], movies["Metacritic_norm_round"])
```

Subtract the Fandango_Stars column from the Metacritic_norm_round column, take the absolute value.

```python
movies["fm_diff"] = np.abs(movies["Metacritic_norm_round"] - movies["Fandango_Stars"])
```

Sort movies based on the difference to find the movies with the largest differences between their Metacritic and Fandango ratings.

```python
movies.sort_values(by="fm_diff", ascending=True).head(5)
```

## 5. Fandango and Metacritic correlation

Let's check the correlation coefficient between Fandango_Stars and Metacritic_norm_round to determine if Fandango consistently has higher scores than Metacritic, or if only a few movies were assigned higher ratings.

```python
from scipy.stats import pearsonr
pearsonr(movies["Fandango_Stars"], movies["Metacritic_norm_round"])
```

The correlation is very low. This would mean Fandango consistently has higher scores then Metacritic

Let's create a linear regression with Metacritic_norm_round as the x values and Fandango_Stars as the y values

```python
from scipy.stats import linregress
slope, intercept, r_value, p_value, std_err = linregress(movies["Metacritic_norm_round"], movies["Fandango_Stars"])
slope, intercept, r_value, p_value, std_err
```

```python
predict_1 = 1 * slope + intercept
predict_1

predict_5 = 5 * slope + intercept
predict_5

plt.scatter(movies["Metacritic_norm_round"], movies["Fandango_Stars"])
plt.plot([1.0, 5.0], [predict_1, predict_5])
plt.xlim(1,5)
plt.show()
```
